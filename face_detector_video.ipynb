## Simple Video Face Detector With MTCNN and CV

```python
from mtcnn import MTCNN
import cv2 as cv
```

```python
video = cv.VideoCapture('angelina.mp4')
detector_obj= MTCNN()

while True:
    success , img = video.read() #success value to make sure it's reading the frames successfully.
    
#     img = cv.resize(img , (100 , 300))
    rgb_img = cv.cvtColor(img , cv.COLOR_BGR2RGB)
    face = detector_obj.detect_faces(rgb_img)[0]
    
    x , y , w , h = face['box']
    conf = face['confidence']

    cv.rectangle(img , (x, y ) , (x+ w , y+ h) , (0,255,0) , 2)

    cv.circle(img ,(216, 210) , 1 ,(0,255,0), 1 )
    
    cv.putText(img , f"CF : {conf:0.2}" , (x , y - 10)  , cv.FONT_HERSHEY_PLAIN,  1 , (255,255,0) , 1 )

    for i , item in face['keypoints'].items():
        cv.circle(img ,item , 5 ,(0,255,0),2 )
        
    cv.imshow('MyWin',img)
    if cv.waitKey(30) == ord('q'):
        break
cv.destroyAllWindows()
```
